```python
import torch
import torchaudio as ta
from chatterbox.tts import ChatterboxTTS
from google.colab import files
import os
from IPython.display import Audio, display

# Check for GPU availability
if torch.cuda.is_available():
    device = "cuda"
    print("✅ GPU is available. Using CUDA.")
else:
    device = "cpu"
    print("⚠️ GPU not available. Using CPU. This might be slow.")

print("✅ Imports and device check complete.")
```

```python
try:
    model = ChatterboxTTS.from_pretrained(device=device)
    print("✅ ChatterboxTTS model loaded successfully.")
except Exception as e:
    print(f"❌ Error loading ChatterboxTTS model: {e}")
    print("Ensure the model name/path is correct and that the library is properly installed.")
    model = None # Set model to None if loading fails
```

```python
if model:
    text_default = "Ezreal and Jinx teamed up with Ahri, Yasuo, and Teemo to take down the enemy's Nexus in an epic late-game pentakill."
    print(f"\nGenerating speech for: '{text_default}' with the default voice...")

    try:
        wav_default = model.generate(text_default)
        output_path_default = "test_default_voice.wav"
        ta.save(output_path_default, wav_default, model.sr)
        print(f"✅ Default voice audio saved to {output_path_default}")
        display(Audio(output_path_default, autoplay=False))
    except Exception as e:
        print(f"❌ Error generating speech with default voice: {e}")
else:
    print("Skipping default voice generation as model failed to load.")
```

```python
# Upload your audio prompt file (e.g., chunk.wav)
print("Please upload your audio prompt file (e.g., 'chunk.wav') for voice cloning.")
uploaded = files.upload()

audio_prompt_filename = None
if uploaded:
    audio_prompt_filename = list(uploaded.keys())[0]
    # Ensure the file is saved to a known path in Colab, like /content/
    # files.upload() already puts it in /content/ by default if no other path is specified.
    AUDIO_PROMPT_PATH = f"/content/{audio_prompt_filename}"
    print(f"✅ '{audio_prompt_filename}' uploaded and available at {AUDIO_PROMPT_PATH}")
else:
    print("⚠️ No file uploaded. Skipping voice cloning with audio prompt.")
    AUDIO_PROMPT_PATH = None
```

```python
if model and AUDIO_PROMPT_PATH and os.path.exists(AUDIO_PROMPT_PATH):
    text_prompted = "Alright, future innovators, listen up! You’re engineering students in the age of AI—a revolution that’s reshaping the world. This is your moment to seize! Dive into AI with relentless curiosity; it’s not just tech, it’s your ticket to solving problems that matter. Learn fast—consume tutorials, experiment with code, and build projects that scream your potential. Don’t just study AI, create with it—whether it’s a smart app, a predictive model, or a bold idea no one’s thought of yet. And here’s the kicker: establish your digital presence now. Share your projects on GitHub, post your insights on X, and build a portfolio that shouts your name to the world. The future doesn’t wait, and neither should you. Embrace AI, hustle hard, and let your work light up the digital universe. You’ve got this—go make waves!"
    print(f"\nGenerating speech for: '{text_prompted}' using voice from {AUDIO_PROMPT_PATH}...")

    try:
        wav_prompted = model.generate(text_prompted, audio_prompt_path=AUDIO_PROMPT_PATH)
        output_path_prompted = "test_custom_voice.wav"
        ta.save(output_path_prompted, wav_prompted, model.sr)
        print(f"✅ Custom voice audio saved to {output_path_prompted}")
        display(Audio(output_path_prompted, autoplay=False))
    except Exception as e:
        print(f"❌ Error generating speech with audio prompt: {e}")
elif not model:
    print("Skipping custom voice generation as model failed to load.")
elif not AUDIO_PROMPT_PATH:
    print("Skipping custom voice generation as no audio prompt was uploaded.")
elif not os.path.exists(AUDIO_PROMPT_PATH):
    print(f"Skipping custom voice generation as the audio prompt file was not found at {AUDIO_PROMPT_PATH}.")
```
